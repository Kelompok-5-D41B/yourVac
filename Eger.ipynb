```python
from selenium import webdriver
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
import json
```

```python
import re
```

```python
import urllib.request
```

```python
url = "https://eigerindostore.com/eiger/mountaineering/men/bags-packs.html?product_list_limit=30"
```

```python
driver = webdriver.PhantomJS()
```

```python
driver.implicitly_wait(10)
```

```python
driver.get(url)
```

```python
print(driver.current_url)
```

```python
for product in driver.find_elements_by_xpath("//div[@class='product-item-info']"):
    print(product.text)
```

```python
products = []
```

```python
for product in driver.find_elements_by_xpath("//div[@class='product-item-info']"):
    if product.text != '':
        details = re.split('\n',product.text)
        images = []
        for img in product.find_elements_by_tag_name('img'):
            images.append({'alt': img.get_attribute('alt'), 'src': img.get_attribute('src')})
        if (len(details) == 4):
             products.append(
             {
                 'Diskon':details[0],
                 'Model': details[1],
                 'Harga': details[2],
                 'Harga_Diskon': details[3],
                 'brand_img': images[0]['src'],
             })
        else:
            products.append(
             {
                 'Model': details[0],
                 'Harga': details[1],
                 'brand_img': images[0]['src'],
             })
                
```

```python
products
```

```python
with open('Eiger.json', 'w') as file:
    json.dump(products, file, indent=4)
```
